EDGE SPLITTING

```python
from torch_sparse import SparseTensor
from torch_geometric.utils import to_undirected, add_self_loops, negative_sampling
import torch
import numpy as np

# Function for edge splitting with path masking
def edge_split_with_path_masking(data, mask_ratio=0.35, val_ratio=0.1, test_ratio=0.2):
    """
    Splits the edges into train/validation/test sets and applies path masking on training edges.

    Args:
        data: A PyTorch Geometric Data object.
        mask_ratio (float): Ratio of training edges to mask.
        val_ratio (float): Ratio of edges for validation set.
        test_ratio (float): Ratio of edges for test set.

    Returns:
        Updated data object with train/val/test splits and masked edges.
    """
    edge_index, _ = add_self_loops(to_undirected(data.edge_index), num_nodes=data.num_nodes)

    # Shuffle edges
    perm = torch.randperm(edge_index.size(1))
    edge_index = edge_index[:, perm]

    # Split edges into train/val/test
    num_edges = edge_index.size(1)
    num_val = int(num_edges * val_ratio)
    num_test = int(num_edges * test_ratio)
    num_train = num_edges - num_val - num_test

    train_edge_index = edge_index[:, :num_train]
    val_edge_index = edge_index[:, num_train:num_train + num_val]
    test_edge_index = edge_index[:, num_train + num_val:]

    # Apply path masking for training edges
    num_mask = int(train_edge_index.size(1) * mask_ratio)
    mask_indices = torch.randperm(train_edge_index.size(1))[:num_mask]
    train_mask_edge_index = train_edge_index[:, mask_indices]
    train_unmasked_edge_index = train_edge_index[:, num_mask:]

    # Sparse adjacency matrix for masked training edges
    edge_index_unmasked = to_undirected(train_unmasked_edge_index)
    edge_index_masked = to_undirected(train_mask_edge_index)
    edge_index_with_loops, _ = add_self_loops(edge_index_unmasked, num_nodes=data.num_nodes)
    adj_sparse = SparseTensor.from_edge_index(edge_index_with_loops).t()

    # Assign edges to data
    data.train_pos_edge_index = train_unmasked_edge_index
    data.train_mask_edge_index = train_mask_edge_index
    data.val_pos_edge_index = val_edge_index
    data.test_pos_edge_index = test_edge_index

    # Generate negative samples
    data.train_neg_edge_index = negative_sampling(
        edge_index=train_unmasked_edge_index,
        num_nodes=data.num_nodes,
        num_neg_samples=train_unmasked_edge_index.size(1),
    )
    data.val_neg_edge_index = negative_sampling(
        edge_index=val_edge_index,
        num_nodes=data.num_nodes,
        num_neg_samples=val_edge_index.size(1),
    )
    data.test_neg_edge_index = negative_sampling(
        edge_index=test_edge_index,
        num_nodes=data.num_nodes,
        num_neg_samples=test_edge_index.size(1),
    )

    return data, adj_sparse

# Apply the updated edge splitting logic
data, adj_sparse = edge_split_with_path_masking(data, mask_ratio=0.35, val_ratio=0.1, test_ratio=0.2)




# Print edge counts
print(f"Training Positive Edges: {data.train_pos_edge_index.size(1)}")
print(f"Training Masked Edges: {data.train_mask_edge_index.size(1)}")
print(f"Validation Positive Edges: {data.val_pos_edge_index.size(1)}")
print(f"Testing Positive Edges: {data.test_pos_edge_index.size(1)}")
print(f"Training Negative Edges: {data.train_neg_edge_index.size(1)}")
print(f"Validation Negative Edges: {data.val_neg_edge_index.size(1)}")
print(f"Testing Negative Edges: {data.test_neg_edge_index.size(1)}")
```

PATH MASKING USING K HOP

```python
# Path masking using k-hop subgraph
def apply_path_masking(edge_index, num_nodes, path_length=3, retain_prob=0.95):
    mask = torch.bernoulli(torch.full((edge_index.size(1),), retain_prob)).bool()
    masked_edge_index = edge_index[:, mask]
    paths_to_mask = []

    valid_nodes = torch.arange(num_nodes)
    for _ in range(path_length):
        center_node = valid_nodes[torch.randint(0, len(valid_nodes), (1,))]
        try:
            nodes, edge_index_subgraph, _, path_edges = k_hop_subgraph(
                node_idx=center_node.item(),
                num_hops=path_length,
                edge_index=edge_index,
                relabel_nodes=False,
            )
            if path_edges.size(1) > 0:
                paths_to_mask.append(path_edges)
        except IndexError:
            continue

    if paths_to_mask:
        paths_to_mask = torch.cat(paths_to_mask, dim=1)
    else:
        paths_to_mask = torch.empty((2, 0), dtype=torch.long)

    return masked_edge_index, paths_to_mask
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

def visualize_graph(edge_index, num_nodes, title="Graph Visualization", masked_edges=None, mask_ratio=None):
    """
    Visualizes a graph using NetworkX.
    
    - edge_index: Graph edges in PyG format.
    - num_nodes: Number of nodes.
    - title: Title of the plot.
    - masked_edges: Edges that were masked (if any).
    - mask_ratio: Percentage of edges masked (optional).
    """
    plt.figure(figsize=(9, 7))

    # Convert to NetworkX Graph
    G = nx.Graph()
    G.add_edges_from(edge_index.t().cpu().numpy())

    # Use force-directed layout for clarity
    pos = nx.spring_layout(G, seed=42, k=1.0)

    # Sample edges for readability
    sampled_edges = random.sample(G.edges(), min(len(G.edges()), 300))  # Max 300 edges
    G_sampled = nx.Graph()
    G_sampled.add_edges_from(sampled_edges)

    # Draw Base Graph (Unmasked)
    nx.draw(G_sampled, pos, node_color='lightblue', edge_color='gray',
            with_labels=False, node_size=100, alpha=0.6)

    # Highlight Masked Edges in Red
    if masked_edges is not None:
        masked_G = nx.Graph()
        masked_G.add_edges_from(masked_edges.t().cpu().numpy())
        nx.draw(masked_G, pos, edge_color='red', width=2, alpha=0.6, label="Masked Edges")

    # Add Information Box in Bottom-Right Corner
    info_text = f"Nodes: {num_nodes}\nEdges: {len(G.edges())}"
    if mask_ratio is not None:
        info_text += f"\nMasking: {mask_ratio*100:.1f}%"

    plt.text(1.02, 0.02, info_text, transform=plt.gca().transAxes, fontsize=10,
             verticalalignment='bottom', bbox=dict(facecolor='white', alpha=0.5))

    plt.title(title)
    plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

def visualize_graph(edge_index, num_nodes, title="Graph", masked_paths=None):
    """
    Visualizes the graph while highlighting masked paths.
    
    - `masked_paths`: If provided, these paths will be **highlighted in red**.
    - Retained paths will be in **gray**.
    - Displays **total path count** in the title.
    """
    plt.figure(figsize=(9, 7))

    # Convert edge_index to NetworkX format
    G = nx.Graph()
    G.add_edges_from(edge_index.t().cpu().numpy())

    pos = nx.spring_layout(G, seed=42, k=1.0)  # Fixed layout for consistency

    # Ensure all nodes in masked_paths exist in pos
    if masked_paths is not None:
        masked_paths_np = masked_paths.t().cpu().numpy()
        for u, v in masked_paths_np:
            if u not in pos:
                pos[u] = (random.uniform(-1, 1), random.uniform(-1, 1))
            if v not in pos:
                pos[v] = (random.uniform(-1, 1), random.uniform(-1, 1))

    #  **Step 1: Draw retained paths in light gray (background)**
    nx.draw(G, pos, node_color="lightblue", edge_color="lightgray",
            with_labels=False, node_size=100, alpha=0.6, width=1, label="Retained Paths (Gray)")

    #  **Step 2: Highlight only the masked paths in red**
    if masked_paths is not None:
        masked_G = nx.Graph()
        masked_G.add_edges_from(masked_paths_np)
        nx.draw(masked_G, pos, edge_color="red", width=2, alpha=0.9, label="Masked Paths (Red)")

        #  Display count of masked paths
        plt.text(1.02, 0.05, f"Masked Paths: {masked_paths.size(1)}", 
                 transform=plt.gca().transAxes, fontsize=10,
                 verticalalignment='bottom', bbox=dict(facecolor='white', alpha=0.5))

    #  Display total path count in title
    path_count = edge_index.size(1)
    plt.title(f"{title}\nTotal Paths: {path_count}")
    plt.legend()
    plt.show()


#  **1 Visualize Original Graph**
visualize_graph(data.edge_index, data.num_nodes, title="Original Graph (Before Splitting)")

#  **2 Visualize Graph After Splitting**
visualize_graph(data.train_pos_edge_index, data.num_nodes, title="Graph After Path Splitting")

#  **3 Visualize Graph After Masking (Masked Paths in Red)**
visualize_graph(data.train_pos_edge_index, data.num_nodes, title="Graph After Path Masking", masked_paths=data.train_mask_edge_index)
```

```python
from sklearn.metrics import accuracy_score

def train_model(
    model, data, optimizer, num_epochs=500, patience=50, retain_prob=0.98, path_length=4, kl_weight=0.01
):
    """
    Train the model with early stopping and learning rate scheduler.

    Args:
    - model: The model to be trained.
    - data: Data containing graph structure and features.
    - optimizer: Optimizer for training.
    - num_epochs: Maximum number of epochs.
    - patience: Number of epochs to wait before early stopping.
    - retain_prob: Probability of retaining edges during masking.
    - path_length: Length of the paths for edge masking.
    - kl_weight: Weight of the KL loss term.

    Returns:
    - model: Trained model.
    - train_losses: List of training losses.
    - val_losses: List of validation losses.
    - val_accuracies: List of validation accuracies.
    - train_accuracies: List of training accuracies.
    """
    model.train()
    best_val_loss = float('inf')
    epochs_no_improve = 0
    train_losses, val_losses, val_accuracies, train_accuracies = [], [], [], []

    # Learning rate scheduler
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=10, verbose=True)

    for epoch in range(1, num_epochs + 1):
        optimizer.zero_grad()

        # Apply path masking
        masked_edge_index, _ = apply_path_masking(
            data.train_pos_edge_index, data.num_nodes, path_length, retain_prob
        )

        # Forward pass
        z = model(data.x, masked_edge_index)
        recon_loss = model.recon_loss(z, data.train_pos_edge_index, data.train_neg_edge_index)
        kl_loss = model.kl_loss()
        total_loss = recon_loss + kl_weight * kl_loss
        total_loss.backward()

        # Gradient clipping
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        train_losses.append(total_loss.item())

        # Calculate training accuracy
        model.eval()
        with torch.no_grad():
            train_z = model(data.x, data.train_pos_edge_index)
            train_pos_scores = model.decode(train_z, data.train_pos_edge_index)
            train_neg_scores = model.decode(train_z, data.train_neg_edge_index)
            train_y_true = torch.cat([torch.ones(train_pos_scores.size(0)), torch.zeros(train_neg_scores.size(0))])
            train_y_pred = torch.cat([train_pos_scores, train_neg_scores])
            train_accuracy = accuracy_score(train_y_true.numpy(), (train_y_pred.numpy() > 0.5).astype(float))
            train_accuracies.append(train_accuracy)

            # Validation step
            z_val = model(data.x, data.val_pos_edge_index)
            val_recon_loss = model.recon_loss(z_val, data.val_pos_edge_index, data.val_neg_edge_index)
            val_losses.append(val_recon_loss.item())

            val_pos_scores = model.decode(z_val, data.val_pos_edge_index)
            val_neg_scores = model.decode(z_val, data.val_neg_edge_index)
            val_y_true = torch.cat([torch.ones(val_pos_scores.size(0)), torch.zeros(val_neg_scores.size(0))])
            val_y_pred = torch.cat([val_pos_scores, val_neg_scores])
            accuracy = accuracy_score(val_y_true.numpy(), (val_y_pred.numpy() > 0.5).astype(float))
            val_accuracies.append(accuracy)

        # Update scheduler
        scheduler.step(val_recon_loss)

        print(
            f"Epoch {epoch}/{num_epochs}, "
            f"Train Loss: {total_loss.item():.4f}, "
            f"Train Accuracy: {train_accuracy:.4f}, "
            f"Val Loss: {val_recon_loss.item():.4f}, "
            f"Val Accuracy: {accuracy:.4f}"
        )

        # Early stopping based on validation loss
        if val_recon_loss < best_val_loss:
            best_val_loss = val_recon_loss
            epochs_no_improve = 0
        else:
            epochs_no_improve += 1

        if patience and epochs_no_improve >= patience:
            print("Early stopping triggered.")
            break

        model.train ()

    return model, train_losses, val_losses, val_accuracies, train_accuracies
```

```python
def find_optimal_threshold(y_true, y_pred):
    thresholds = np.linspace(0, 1, 100)
    f1_scores = [f1_score(y_true, (y_pred > t).astype(int)) for t in thresholds]
    best_threshold = thresholds[np.argmax(f1_scores)]
    return best_threshold

def test_model_with_threshold(model, data):
    model.eval()
    with torch.no_grad():
        # Get latent embeddings
        z = model(data.x, data.test_pos_edge_index)

    # Decode edges
    pos_scores = model.decode(z, data.test_pos_edge_index)
    neg_scores = model.decode(z, data.test_neg_edge_index)

    # Prepare true labels and predictions
    y_true = torch.cat([torch.ones(pos_scores.size(0)), torch.zeros(neg_scores.size(0))]).cpu().numpy()
    y_pred = torch.cat([pos_scores, neg_scores]).cpu().numpy()

    # Find optimal threshold using F1 score
    thresholds = np.linspace(0, 1, 100)
    f1_scores = [f1_score(y_true, (y_pred > t).astype(int)) for t in thresholds]
    optimal_threshold = thresholds[np.argmax(f1_scores)]
    print(f"Optimal Threshold: {optimal_threshold}")

    # Binarize predictions using the optimal threshold
    y_pred_binary = (y_pred > optimal_threshold).astype(float)

    # Calculate metrics
    auc = roc_auc_score(y_true, y_pred)
    ap = average_precision_score(y_true, y_pred)
    accuracy = accuracy_score(y_true, y_pred_binary)
    precision = precision_score(y_true, y_pred_binary)
    recall = recall_score(y_true, y_pred_binary)
    f1 = f1_score(y_true, y_pred_binary)

    print(f"Test Results - AUC: {auc:.4f}, AP: {ap:.4f}")
    print(f"Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}")

    return {
        "AUC": auc,
        "AP": ap,
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1,
        "Optimal Threshold": optimal_threshold
    }

# Evaluate the model with dynamic threshold
metrics = test_model_with_threshold(trained_model, data)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import torch

def plot_auc_roc_curve(model, data, plot_title="ROC Curve", figsize=(8, 6), color="blue", diagonal=True):
    """
    Plots the Receiver Operating Characteristic (ROC) curve for a trained model.

    Args:
    - model: Trained model object.
    - data: Data object containing test edge indices and node features.
    - plot_title (str): Title of the plot.
    - figsize (tuple): Figure size for the plot.
    - color (str): Line color for the ROC curve.
    - diagonal (bool): Whether to plot the diagonal reference line.

    Returns:
    - auc_score (float): Computed AUC score.
    """
    # Set model to evaluation mode
    model.eval()
    with torch.no_grad():
        # Get latent embeddings
        z = model(data.x, data.test_pos_edge_index)

    # Decode edges
    pos_scores = model.decode(z, data.test_pos_edge_index)
    neg_scores = model.decode(z, data.test_neg_edge_index)

    # Prepare true labels and predictions
    y_true = torch.cat([torch.ones(pos_scores.size(0)), torch.zeros(neg_scores.size(0))]).cpu().numpy()
    y_pred = torch.cat([pos_scores, neg_scores]).cpu().numpy()

    # Compute ROC curve and AUC
    fpr, tpr, _ = roc_curve(y_true, y_pred)
    auc_score = roc_auc_score(y_true, y_pred)

    # Plot the ROC curve
    plt.figure(figsize=figsize)
    plt.plot(fpr, tpr, label=f'AUC = {auc_score:.4f}', color=color)
    if diagonal:
        plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label="Random Guess")
    plt.title(plot_title)
    plt.xlabel('False Positive Rate (FPR)')
    plt.ylabel('True Positive Rate (TPR)')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.show()

    return auc_score

# Example Usage
auc = plot_auc_roc_curve(trained_model, data)
print(f"Computed AUC: {auc:.4f}")
```
